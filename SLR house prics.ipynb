```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('prediction.csv')
```

```python
df
```

```python
%matplotlib inline
plt.scatter(df.area,df.prices,color = 'blue', marker = '+' )
plt.xlabel('area')
plt.ylabel('price')
plt.show()
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']],df.prices)
```

```python
reg.predict([[3000]])
```
